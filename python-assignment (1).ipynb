```python
#32 Pythagorian triplets
lim=60
a,b,c=0,0,0
triplets=[]
m=1
while c<lim:
    n=m+1
    a=n*n-m*m
    b=2*n*m
    c=m*m+n*n
    m+=1
    if c>lim:
        break
    triplets.append((a,b,c))
print(triplets)
```

```python
# 36 sum of digits of factorial of numbers from 1 to 5
# step 1->Factorial using recursion and memory optimization
#step 2->sum of digit
memo={}
res=0
def factorial(n):
    if n in memo:
        return memo[n]
    if n==0:
        return 1
    result =n*factorial(n-1)
    memo[n]=result
    return result

def sumofdigit(i):
    add=0
    while i>0:
        digits=i%10
        add=add+digits
        i=i//10
    return add

factorial(5)
for i in memo:
    res=res+sumofdigit(memo[i])
print(res)
    

```
